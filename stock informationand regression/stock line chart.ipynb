```python
import matplotlib.pyplot as plt
import seaborn as sns
import seaborn.regression as snsl

import datetime
import tushare as ts
```

## The closing price curve of stock

```python
code = '600678'
end = datetime.date.today()
#开始时间结束时间，选取最近一年的数据
start = end - datetime.timedelta(days=365)

end = end.strftime('%Y-%m-%d')
start = start.strftime('%Y-%m-%d')

stock = ts.get_hist_data(code,start,end)#选取一支股票
stock['close'].plot(legend=True ,figsize=(10,4))
```

## 5days MA[moving average]、10days MA,20days MA and the close price

```python
stock[['close','ma5','ma10','ma20']].plot(legend=True ,figsize=(10,4))
plt.savefig(code + ".png")
```

## daliy percantage changes

```python
stock['Daily Return'] = stock['close'].pct_change()
stock['Daily Return'].plot(legend=True,figsize=(10,4))
```
